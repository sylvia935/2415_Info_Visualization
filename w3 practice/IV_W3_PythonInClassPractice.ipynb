```python
x = np.random.randn(50)
```

```python
y = np.random.randn(50) + 3*x
```

```python
def OLSRegressPlot(x,y,col,xlab,ylab):
    xx = sm.add_constant(x, prepend=True)
    res = sm.OLS(y,xx).fit()
    constant, beta = res.params
    r2 = res.rsquared
    lab = r'$slope = %.2f, \,R^2 = %.2f$' %(beta,r2)
    plt.scatter(x,y,s=30,facecolors='none', edgecolors=col)
    plt.plot(x,constant + x*beta,"red",label=lab)
    plt.legend(loc = 'best',fontsize=16)
    plt.xlabel(xlab,size=16)
    plt.ylabel(ylab,size=16)

fig = plt.figure(figsize=(7, 7),facecolor='white')
OLSRegressPlot(x,y,'RoyalBlue',r'$x$',r'$y$')
plt.show()
```

```python
data = norm.rvs(10.0, 2.5, size=5000)
```

```python
mu, std = norm.fit(data)
```

```python
fig = plt.figure(figsize=(7, 7),facecolor='white')
plt.hist(data, bins=40, normed=True, alpha=0.6, color='g') #alpha ->transparency 
x = np.linspace(min(data), max(data), 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'r', linewidth=3)  #drawing continuous line 
title = r"$\mu = %.2f, \,  \sigma = %.2f$" % (mu, std)
plt.title(title,size=16)
plt.show()
```
